> apply()
>- axis가 0 or 'index'인 경우 각 column에 대해 함수를 적용합니다  
>- axis가 1 or 'columns'인 경우 각 row에 대해 함수를 적용합니다

```python
df = pd.DataFrame([[4,9]] * 3, columns=["A", "B"])
df
```

```python
df.apply(np.sqrt)
```

```python
df.apply(np.sum, axis=0)
```

```python
df.apply(np.sum, axis=1)
```

```python
df.apply(lambda x: [1, 2], axis=0)
```

```python
df.apply(lambda x: [1, 2], axis=1)
```

```python
df.apply(lambda x: [1, 2], axis=1, result_type='expand')
```

```python
df.apply(lambda x: [1, 2], axis=1, result_type='broadcast')
```

```python
df3 = pd.DataFrame({'A': [1,3,4,3,4],'B':[2,3,1,2,3],'C':[1,5,2,4,4]})
df3
```

```python
df3.apply(lambda x: x.max() - x.min()) #각 열에 대한 최대값-최솟값
```

```python
df3.apply(lambda x: x.max() - x.min(), axis=1) 
```

```python
df3.apply(pd.value_counts) # 각 column에 대해 어떤값이 사용되는 지 알려면 value_counts
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic["adult/child"] = titanic.apply(lambda x: "adult" if x.age >= 20 else "child", axis=1)
titanic.tail()
```

```python
titanic["category1"] = titanic.apply(lambda x: x["sex"] if x.age >= 20 else "child", axis=1)
titanic
```
