# 범례(Legend)
>여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료룰 
표시하는지를 보여주기 위해 legend 명령으로 범례를 추가할 수 있습니다.  
범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면  
loc 키워드 인수를 사용합니다.

> 

```python
X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls = ":", label= "sine")
plt.legend(loc=2)
plt.show()
```

```python
# x,y축 label 각 축들이 의미하는 바를 표시
X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X), np.sin(X)
plt.title("Cosine Plot")
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("time")

plt.show()
```

# Figure
> Matplotlib는 data를 Figur에 x-y좌표의 그래프로 표시합니다. 각각의 Figure는 하나 이상의 Axes를 갖습니다.  

>가장 간단하게 Figure, Axes 객체를 만드는 방법은 pyplot.subplots()을 사용하는 것  
그 후 Axes.plot()를 활용하여 Axes에 데이터를 그립니다.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3]);
```

```python
fig = plt.figure() # an empty figure with no Axes
fig, ax = plt.subplots() # a figure with a single Axes
fig, axs = plt.subplots(2,2) # a figure with a 2x2 qrid of Axes
```

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10,2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

>주피터는 내부에 ipython이 돌고 있다. show()메서드를 통해 그래프를 그려보면 window 객체를 생성하지 않고 주피터 위젯을 통해 보여줌

```python
# 현재 사용하고 있는 Figure객체를 얻으려면(다른 변수에 할당할 수도 있음).gcf()메서드를 사용합니다.
f1 = plt.figure(1)
plt.title("현재의 Figure 객체")
plt.plot([1,2,3,4], "ro:")

f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.show()
```
